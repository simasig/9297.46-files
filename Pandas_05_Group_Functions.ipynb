# Pandas DataFrame agg() Method

#Pandas Series.agg() is used to pass a function or list of function to be applied on a series or even each element of series separately.

```python
pup.groupby("Country")[["income"]].mean()
```

```python
pup.groupby("Country")[["income"]].median()
```

```python
pup.groupby("Country")[["income"]].agg("mean")
```

```python
pup.groupby("Country")[["income"]].agg(["mean","count","median"])
```

```python
pup.agg('count').to_frame()
```

```python
pup.groupby("Country").agg(['count','max'])
```

```python
pup.groupby("Country")['income'].agg(['count','max','sum'])
```

```python
pup.groupby("Country")[['income','rooms']].agg(['count','max','sum'])
```

```python
#Different aggrigation per column
```

```python
pup.groupby("Country").agg({'Name':'count','income':['sum','mean','max'],'Age':'max','rooms':'mean'})
```

```python
pup.groupby("Country").agg(
    count_name=('Name','count'),
    mean_income=('income','mean'),
    max_age = ('Age','max'))
```
